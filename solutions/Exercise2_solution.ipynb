The equation we were given was
\begin{equation}
d = 2r\arcsin \Bigg( \sqrt{\sin^2\bigg(\frac{\phi_2-\phi_1}{2}\bigg) + \cos{(\phi_1)}\cos{(\phi_2)}\sin^2\bigg(\frac{\lambda_2-\lambda_1}{2}\bigg)}  \Bigg).
\end{equation}

and we wanted to find the distance between Granchester Tea Rooms ($\phi=52.177364$, $\lambda=0.096534$) and Great St Mary's ($\phi=52.205301$, $\lambda=0.118262$).

I will write a single cell block as the function and another cell block to run the code.  To understand the answer read the comment lines in the code and do ask for help where sections are unclear.

```python
#imports
# we will need numpy for trig functions
import numpy as np

#let's define our function
# in the simplest implementation it is going to take four separate variables, the lat and lon's for two points
# on the surface of the Earth.  But, you could package all this information into a single list, or many altneratives.
def geo_dist(lat1, lat2, lon1, lon2):
    #first we need to convert from degrees, which presumably our input is in, to radians
    lt1 = lat1/180 * np.pi
    lt2 = lat2/180 * np.pi
    ln1 = lon1/180 * np.pi
    ln2 = lon2/180 * np.pi

    #we will need to hardcode the radius of the sphere (Earth), the units we decide on here
    # set the units for the output
    r = 6371.
    
    #let's break the haversine formula up into smaller chunks and piece them altogether at the end.
    A = np.sin((lt2 - lt1)/2)**2
    B = np.cos(lt1)*np.cos(lt2)*np.sin((ln2 - ln1)/2)**2
    
    return 2*r*np.arcsin((A + B)**0.5)
```

```python
#define the two locations
lat1 = 52.177364; lon1 = 0.096534 #Granchester
lat2 = 52.205301; lon2 = 0.118262 #Gt St Mary's

#run our function and save the output
dist = geo_dist(lat1, lat2, lon1, lon2)

#print the result, formatting the output to limit the number of decmial places
print('The distance from Gt St Mary\'s to Granchester tea rooms is', '{:.3f}'.format(dist), 'km')
```
