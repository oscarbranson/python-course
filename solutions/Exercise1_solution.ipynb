## Predator prey interaction

<div class=obj>
    <b>Aim:</b> Write a while loop to calculate 100 years of predator prey interactions.
</div>



```python
#let's define our variables
# first for the trilobites
gx = 1.1 #growth
dx = 0.4 #death
x0 = 10  #initial population
# now for the nautiloids
gy = 0.1 #growth
dy = 0.4 #death
y0 = 10  #initial population

#setup the lists we are going to store all the results in
x = [x0]
y = [y0]
t = [0]

#time step
dt = 0.01
#maximum time to run our calculation until
tmax = 100

#start our loop
while t[-1] < tmax:
    #append our new time
    t.append(t[-1] + dt)
    x.append( x[-1] + (gx*x[-1] - x[-1]*y[-1]*dx)*dt )
    y.append( y[-1] + (gy*x[-1]*y[-1] - y[-1]*dy)*dt )

import matplotlib.pyplot as plt

plt.plot(t,y,label="Nautiloids")
plt.plot(t,x,label="Trilobites")
plt.xlabel('time (yr)');
plt.ylabel('population size');
plt.legend()
```
