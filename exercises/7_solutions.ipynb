## Installing PhreeQC

1. Go to the [PhreeQC website](https://www.usgs.gov/software/phreeqc-version-3).
2. Scroll down to the `iPhreeQC Modules` section, and download the appropriate version for your computer. Save this somewhere sensible.
3. Unpack the package you've just download, and follow ths instructions in the 'Basic Installation' section of the `INSTALL` file.


```python
import pandas as pd
import phreeqpy.iphreeqc.phreeqc_dll as phreeqc_com
```

```python
phreeqc = phreeqc_com.IPhreeqc('/usr/local/lib/libiphreeqc.so')

phreeqc.load_database("/home/oscar/phreeqc/iphreeqc-3.8.6-17100/database/phreeqc.dat")

input_str ="""
SOLUTION 1
    pH      8.5
    pe      4.0
    temp    25.0
    units   mmol/kgw
    density 1.0
    Ca      10.0
    Mg      5.0
    Na      20.0
    K       5.0
    Cl      30.0
    SO4     15.0
    C(4)    2.0
SELECTED_OUTPUT
    -reset   false
    -pH      true
    -pe      true
    -totals  Ca Mg Na K Cl S C
    -activities Ca+2 Mg+2 Na+ K+ Cl- SO4-2 HCO3- CO3-2
    -si      calcite
END
"""

phreeqc.run_string(input_str)
```

```python
out = phreeqc.get_selected_output_array()
solution = pd.DataFrame(out[1:], columns=out[0])

solution
```
